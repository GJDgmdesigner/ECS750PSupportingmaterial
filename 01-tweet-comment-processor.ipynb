```python
import numpy as np
import pandas as pd
import json
```

This part is to process the downloaded JSON twitter comment, settle the data to the dataframe

```python
# The user need to manually chage the date name(change the filename) to process all raw json data
date = '06_' + '20'                                                       #change the date name to process different date comments
filename = date + '.json'
fileaddress = './callofdutytwittercomment/' + str(filename)
#comment_data = json.load(open('/kaggle/input/callofdutytwittercomment/06_14.json'))
comment_data = json.load(open(fileaddress))
comment_df = pd.DataFrame(comment_data["data"])
entity = comment_df["entities"]
public_metrics = comment_df["public_metrics"]
print(len(comment_df))
comment_list_length = len(comment_df)        #the length of dataframe
display(comment_df)
#print(entity)

unknown_index_list = []

#print(entity.isnull())


# Mark out the unknown author tweers
for x in range(comment_list_length):        #some comment cannot find author, pick them up
    if(entity.isnull()[x] == True):
        unknown_index_list.append(x)
    
print(unknown_index_list)
```

```python
# Classify all data into lists
username_list = []
retweet_list = []
reply_list = []
like_list = []
quote_list = []
bookmark_list = []
impression_list = []


for x in range(comment_list_length):
    if x in unknown_index_list:
        username = "**this tweet had been deleted or hidden**"
    else:
        username = entity[x]['mentions'][0]['username']
    
    retweet = public_metrics[x]['retweet_count']
    reply = public_metrics[x]['reply_count']
    like = public_metrics[x]['like_count']
    quote = public_metrics[x]['quote_count']
    bookmark = public_metrics[x]['bookmark_count']
    impression = public_metrics[x]['impression_count']
    username_list.append(username)
    retweet_list.append(retweet)
    reply_list.append(reply)
    like_list.append(like)
    quote_list.append(quote)
    bookmark_list.append(bookmark)
    impression_list.append(impression)
```

```python
# Create dataframe columns
comment_df['username'] = username_list
comment_df['retweet'] = retweet_list
comment_df['reply'] = reply_list
comment_df['like'] = like_list
comment_df['quote'] = quote_list
comment_df['bookmark'] = bookmark_list

comment_df.drop(['edit_history_tweet_ids'], axis = 1,inplace=True)
#comment_df.drop(['author_id'], axis = 1,inplace=True)
comment_df.drop(['id'], axis = 1,inplace=True)
comment_df.drop(['entities'], axis = 1,inplace=True)
comment_df.drop(['public_metrics'], axis = 1,inplace=True)
```
